## Question 1
Print all elements of a tuple using a for loop.

```python
t = (1, 2, 3, 4, 5)
for item in t:
    print(item)
```

## Question 2
Print all elements of a tuple using a while loop.

```python
t = (6,7,8,9,10)
i = 0
while i < len(t):
    print(t[i])
    i += 1
```

## Question 5
Count the number of even and odd elements in a tuple.

```python
u = (1, 2, 3, 4, 5, 6, 7, 8)
even = 0
odd = 0         
for num in u:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1
print(even)
print(odd)
```

## Question 13
Check whether a tuple is a palindrome or not.

```python
t = (1, 2, 3, 2, 1)

if t == t[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```
