```python
#Program 1: Check Whether a Number is Positive or Negative (Using Simple if Statement)
num = int(input("Enter a number: "))
if num >= 0:
    print("Positive no")
if num < 0:
    print("Negative no")
```

```python
#Write a Python program to check whether the given number is even or odd using the modulus (%) operator.
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
#Write a Python program to check whether the given number is even or odd using a bitwise operator.
num = int(input("Enter a number: "))
if num & 1:
    print("Odd Number")
else:
    print("Even Number")
```

```python
#Python Program to Check if Two Numbers are Equal or Not
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
if num1 == num2:
    print("Both numbers are equal")
else:
    print("Both numbers are not equal")
```

```python
#Write a Python program to check if two numbers are equal or not using a bitwise operator.
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
if (num1 ^ num2) == 0:
    print("Both numbers are equal")
else:
    print("Both numbers are not equal")
```

```python
#Write a Python program to find the largest number among two numbers.
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
if num1 > num2:
    print("Largest number =", num1)
else:
    print("Largest number =", num2)
```

```python
#Write a Python program to find the smallest number among two numbers.
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
if num1 < num2:
    print("Smallest number =", num1)
else:
    print("Smallest number =", num2)
```

```python
#Write a Python program to find the maximum and minimum of two numbers without using any loop or condition. (Use the abs() function.)
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
maximum = (num1 + num2 + abs(num1 - num2)) / 2
minimum = (num1 + num2 - abs(num1 - num2)) / 2
print("Maximum =", maximum)
print("Minimum =", minimum)
```
