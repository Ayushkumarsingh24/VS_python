```python
# factorial Caluculation (Reverse Display)
def factorial_with_recursion(n):
    if n == 1:
        return 1
    return n * factorial_with_recursion(n - 1)

print(factorial_with_recursion(2))
print(factorial_with_recursion(3))
print(factorial_with_recursion(4))
print(factorial_with_recursion(5))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = 17
print(f"{num} is prime:", is_prime(num))
```

## 2. Write function to reverse number

```python
def reverse_number(n):
    return int(str(n)[::-1])

num = 12345
print("Reversed number:", reverse_number(num))
```

## 3. Write function to count vowels

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count

text = "Hello World"
print("Vowel count:", count_vowels(text))
```

## 4. Write function to find max of 3 numbers

```python
def max_of_three(a, b, c):
    return max(a, b, c)

print("Maximum:", max_of_three(10, 25, 15))
```

## 5. Write recursive Fibonacci

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

for i in range(10):
    print(fibonacci(i), end=" ")
```

## 6. Write function using *args to find max

```python
def find_max(*args):
    return max(args)

print("Maximum:", find_max(10, 45, 22, 99, 5))
```

## 7. Write function using **kwargs to display details

```python
def display_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_details(name="John", age=21, city="Delhi")
```

## 8. Write function to check palindrome

```python
def is_palindrome(text):
    return text == text[::-1]

word = "madam"
print(f"{word} is palindrome:", is_palindrome(word))
```

## 9. Write function to calculate average

```python
def calculate_average(*numbers):
    return sum(numbers) / len(numbers)

print("Average:", calculate_average(10, 20, 30, 40))
```

## 10. Write function for calculator

```python
def calculator(a, b, operation):
    if operation == "+":
        return a + b
    elif operation == "-":
        return a - b
    elif operation == "*":
        return a * b
    elif operation == "/":
        return a / b
    else:
        return "Invalid operation"

print("Addition:", calculator(10, 5, "+"))
print("Subtraction:", calculator(10, 5, "-"))
print("Multiplication:", calculator(10, 5, "*"))
print("Division:", calculator(10, 5, "/"))
```
