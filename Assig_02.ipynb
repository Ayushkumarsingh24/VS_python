```python
s = "Hell World"
vowels = consonants = 0
for ch in s.lower():
    if ch.isalpha():
        if ch in "aeiou":
            vowels += 1
        else:
            consonants += 1
print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
s = "nitin"
rev = ""
for ch in s:
    rev = ch + rev
if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
s = "upper"
result = ""
for ch in s:
    if 'a' <= ch <= 'z':
        result += chr(ord(ch) - 32)
    else:
        result += ch
print(result)
```

```python
s = "LOWER"
result = ""
for ch in s:
    if 'A' <= ch <= 'Z':
        result += chr(ord(ch) + 32)
    else:
        result += ch
print(result)
```

```python
s = "Do Shark Sake"
target = "k"
freq = 0
for ch in s:
    if ch == target:
        freq += 1
print(freq)
```

```python
s = "Here We GO Again"
result = ""
for ch in s:
    if ch == " ":
        result += "-"
    else:
        result += ch
print(result)
```

```python
s = "Nausty World"
result = ""
for ch in s:
    if ch.lower() not in "aeiou":
        result += ch
print(result)
```

```python
s1 = "Drop"
s2 = "Boom"
equal = True
if len(s1) != len(s2):
    equal = False
else:
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            equal = False
            break
print("Equal" if equal else "Not Equal")
```

```python
s = "Never33#@"
digits = alphabets = special = 0
for ch in s:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        alphabets += 1
    else:
        special += 1
print("Digits:", digits)
print("Alphabets:", alphabets)
print("Special:", special)
```

```python
s = "Why Lord why So Sick Life"
words = s.split()
longest = words[0]
for w in words:
    if len(w) > len(longest):
        longest = w
print("Longest word:", longest)
```
